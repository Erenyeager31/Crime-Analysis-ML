```python
#! importing libraries
# import numpy as np
import pandas as pd
# import seaborn as sns
# import matplotlib.pyplot as plt
# from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# from sklearn import svm
# from sklearn.metrics import accuracy_score
# from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('ProccessedData.csv')
copydf = pd.read_csv('ProccessedData.csv')
df.head()
```

```python
df = df.drop(['crime description','crime Location','Conclusion','Link/source','Off.Name','Name of victim'],axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Type of crime'].value_counts()
```

```python
df['Date of Offend'] = pd.to_datetime(df['Date of Offend'], errors='coerce')

def replace_invalid_dates(date):
  if pd.isna(date) or not pd.to_datetime(date, errors='coerce'):
    return pd.to_datetime('01-01-2024')
  else:
    return date

# Apply the function to 'Date of Offend' column using vectorized apply
df['Date of Offend'] = df['Date of Offend'].apply(replace_invalid_dates)

print(df)
```

```python
df['Type of crime'].value_counts()
```

```python
df.head()
```

### FEATURE EXTRACTION

```python
le = LabelEncoder()
df = pd.get_dummies(df, columns=['Type of crime'])

# df['Date of Offend'] = pd.to_datetime(df['Date of Offend'])

# Extract relevant features from the date
df['Month'] = df['Date of Offend'].dt.month
df['Day of Week'] = df['Date of Offend'].dt.dayofweek

# Replace 'M' and 'F' with 1 and 0 in 'Off.gender' column
df['Off.gender'].replace({'M': 1, 'F': 0}, inplace=True)

# Replace 'M' and 'F' with 1 and 0 in 'Victim Gender' column
df['Victim Gender'].replace({'M': 1, 'F': 0}, inplace=True)

df['Time_of_Day'] = le.fit_transform(df['Time_of_Day'])
df['Location_Type'] = le.fit_transform(df['Location_Type'])
df['Weather_Conditions'] = le.fit_transform(df['Weather_Conditions'])


# Select features for clustering
features = ['Month', 'Day of Week', 'lat', 'long', 'Victim Gender', 'Victim Age', 'Off.age', 'Off.gender','Time_of_Day','Location_Type','Location_Type'] + list(df.filter(like='Type of crime_').columns)

# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(df[features])

df.head()
```

```python
k = 6

# Apply K-means clustering
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)

# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Explore the clusters
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_df = pd.DataFrame(cluster_centers, columns=features)
print(cluster_df)
```

```python
df['Cluster'].value_counts()
```

### VIEW THE CLUSTER CHARACTERISTICS

```python
# Group data by cluster and calculate mean values for each feature
cluster_characteristics = df.groupby('Cluster').mean(numeric_only=False)
# Round all numeric values
cluster_characteristics = cluster_characteristics.round()

# Convert gender values back to 'M' and 'F'
cluster_characteristics['Off.gender'].replace({1: 'M', 0: 'F'}, inplace=True)
cluster_characteristics['Victim Gender'].replace({1: 'M', 0: 'F'}, inplace=True)

# Print cluster characteristics
cluster_characteristics

cluster_characteristics.head()
```

```python
# Find columns related to 'Type of crime'
type_of_crime_columns = cluster_characteristics.columns[cluster_characteristics.columns.str.startswith('Type of crime_')]

# Find the index of the column with the maximum value for each row
cluster_characteristics['Type of crime'] = cluster_characteristics[type_of_crime_columns].idxmax(axis=1)

# Extract the actual crime type from the column name
cluster_characteristics['Type of crime'] = cluster_characteristics['Type of crime'].str.replace('Type of crime_', '')

# Drop the one-hot encoded columns
cluster_characteristics.drop(columns=type_of_crime_columns, inplace=True)

# Print the DataFrame to verify the changes
cluster_characteristics
```

```python
# Find columns related to 'Type of crime'
type_of_crime_columns = df.columns[df.columns.str.startswith('Type of crime_')]

# Find the index of the column with the maximum value for each row
df['Type of crime'] = df[type_of_crime_columns].idxmax(axis=1)

# Extract the actual crime type from the column name
df['Type of crime'] = df['Type of crime'].str.replace('Type of crime_', '')

# Drop the one-hot encoded columns
df.drop(columns=type_of_crime_columns, inplace=True)
```
