```python

#! Cleaning the data
print(crimeDataset.isnull().sum())

#? replacing all the - with NA
CleanedCrimeDataset = crimeDataset.replace('NA','NA')
CleanedCrimeDataset.fillna('NA',inplace=True)

CleanedCrimeDataset.head()
```

```python
import random

# Function to generate random time of day
def get_random_time_of_day():
    time_of_day = ['Morning', 'Afternoon', 'Evening', 'Night']
    return random.choice(time_of_day)

# Function to generate random location type
def get_random_location_type():
    location_type = ['Residential', 'Commercial', 'Public Park']
    return random.choice(location_type)

# Function to generate random weather conditions
def get_random_weather_conditions():
    weather_conditions = ['Sunny', 'Cloudy', 'Rainy']
    return random.choice(weather_conditions)

# Generate random features
CleanedCrimeDataset['Time_of_Day'] = [get_random_time_of_day() for _ in range(len(CleanedCrimeDataset))]
CleanedCrimeDataset['Location_Type'] = [get_random_location_type() for _ in range(len(CleanedCrimeDataset))]
CleanedCrimeDataset['Weather_Conditions'] = [get_random_weather_conditions() for _ in range(len(CleanedCrimeDataset))]

print(CleanedCrimeDataset)
```
