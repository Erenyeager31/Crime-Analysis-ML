```python

#! filling up missing values for age and gender


#!import the csv files
crimeDataset = pd.read_csv('ProcessedData.csv')
crimeDataset.head()
```

```python
# Iterate through the rows
for index, row in crimeDataset.iterrows():
    # Check if 'Off.age' is a list
    if isinstance(row['Off.age'], list):
        # Convert ages to numeric and calculate average
        ages = [float(age) for age in row['Off.age'] if str(age).replace('.', '', 1).isdigit()]
        if ages:
            avg_age = sum(ages) / len(ages)
            crimeDataset.at[index, 'Off.age'] = avg_age
    
    # Check if 'Off.gender' is a list
    if isinstance(row['Off.gender'], list):
        # Take mode of genders
        genders = [gender for gender in row['Off.gender'] if isinstance(gender, str)]
        if genders:
            mode_gender = max(set(genders), key=genders.count)
            crimeDataset.at[index, 'Off.gender'] = mode_gender

# Check if there are any remaining missing values
missing_values = crimeDataset.isnull().sum()
print(missing_values)
```

```python
import pandas as pd
import numpy as np

# Define genders and age ranges
genders = ['Male', 'Female']
min_age = 18
max_age = 80

# Replace missing values in Victim Gender with random values
crimeDataset['Victim Gender'] = crimeDataset['Victim Gender'].fillna(
    crimeDataset['Victim Gender'].apply(lambda x: np.random.choice(genders) if pd.isna(x) else x))

# Replace missing values in Victim Age with random values
crimeDataset['Victim Age'] = crimeDataset['Victim Age'].fillna(
    crimeDataset['Victim Age'].apply(lambda x: np.random.randint(min_age, max_age + 1) if pd.isna(x) else x))

# Check for remaining missing values
missing_values = crimeDataset.isnull().sum()
print(missing_values)
```
